```python
import tensorflow as tf
import numpy as np
img = tf.random_uniform([10,28,28,3])
sess = tf.Session()
print(sess.run(img))
```

```python
path1 = tf.slice(img,[0,0,0,0],[10,28,28,1])
path2 = tf.slice(img,[0,0,0,1],[10,28,28,2])
path3 = tf.slice(img,[0,0,0,2],[10,28,28,3])
print(path1.shape)
print(path2.shape)
print(path3.shape)
```

```python
shape = tf.shape(img)
rank = tf.rank(shape)
sess = tf.Session()
print(sess.run(rank))
sess = tf.Session()
print(sess.run(shape))
```
